```python
from rgi.rgizero.games.connect4 import Connect4Game
```

```python
import random
game = Connect4Game()
state = game.initial_state()

while not game.is_terminal(state):
    legal_actions = game.legal_actions(state)
    action = random.choice(legal_actions)
    print(f"# Action: {action}")
    state = game.next_state(state, action)

    print(game.pretty_str(state))
    print(f"{'  ' * (action-1)} ^\n")
    
print(f"Winner: {state.winner}")
```

```python
def play_connect4():
    game = Connect4Game()
    state = game.initial_state()
    action0 = None
    while not game.is_terminal(state):
        legal_actions = game.legal_actions(state)
        action = random.choice(legal_actions)
        if action0 is None:
            action0 = action
        state = game.next_state(state, action)
    return state, action0


# Should complete 10k random games quickly
games = [(final_state, action0) for _ in range(10000) for (final_state, action0) in [play_connect4()]]
```

```python
from collections import Counter
counts = Counter((a, state.winner) for (state, a) in games)

for a in range(1, 8):
    total = counts[(a, 1)] + counts[(a, 2)] + counts[(a, 0)]
    print(f'Action {a}: Win/Loss/Draw = {counts[(a, 1)]}/{counts[(a, 2)]}/{counts[(a, 0)]} percent={counts[(a, 1)]/total:6.2%}')
```
