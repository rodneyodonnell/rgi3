# Scratchpad & example notebook for rgizero project.


```python
import os
from pathlib import Path
from importlib import reload

import numpy as np

DATA_DIR = Path.cwd().parent / "data" / "rgizero-scratch"
os.makedirs(DATA_DIR, exist_ok=True)
```

## Create trajectory dataset

```python
from rgi.rgizero.data import trajectory_dataset as td_lib
reload(td_lib)

vocab = td_lib.Vocab(vocab_size=7, itos = '1234567')
td_builder = td_lib.TrajectoryDatasetBuilder(vocab)
for i in range(1000):
    td_builder.add_trajectory(
        actions = np.array([i % 7, 6, 5]),
        policies = np.array(np.random.rand(3, 7)),
        values = np.array([1, 0, -1]),
    )
td_builder.save(DATA_DIR, 'train_1000')
```

```python
td = td_lib.TrajectoryDataset(DATA_DIR, 'train_1000', 5)
td[0]
```

```python
loader = td_lib.build_trajectory_loader(
    DATA_DIR, 'train_1000', block_size=5, batch_size=1,
    device_is_cuda=False, workers=4)

for batch in loader:
    print(batch)

```
